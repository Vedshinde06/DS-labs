```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")


df.head()
```

```python
# Shape of dataset
print("Shape of dataset:", df.shape)

# Display column names
print("\nColumns:")
print(df.columns)

# Display data types
print("\nData Types:")
print(df.dtypes)
```

```python
print("Features and their types:\n")

for column in df.columns:
    print(column, ":", df[column].dtype)
```

```python
# Statistical summary
df.describe()
```

```python
# Set style
sns.set(style="whitegrid")

# Create histograms
df.hist(figsize=(10,8), bins=15)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

```python
# Create boxplots for all numerical columns
plt.figure(figsize=(12,6))

sns.boxplot(data=df.iloc[:,0:4])

plt.title("Boxplot of Iris Dataset Features")
plt.show()
```

```python
# Individual boxplots

features = df.columns[:-1]

for feature in features:
    
    plt.figure(figsize=(6,4))
    
    sns.boxplot(y=df[feature])
    
    plt.title(f"Boxplot of {feature}")
    
    plt.show()
```

```python
# Distribution plots

features = df.columns[:-1]

for feature in features:
    
    plt.figure(figsize=(6,4))
    
    sns.histplot(df[feature], kde=True)
    
    plt.title(f"Distribution of {feature}")
    
    plt.show()
```

```python
# Detect outliers using IQR

features = df.columns[:-1]

for feature in features:
    
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    outliers = df[
        (df[feature] < lower_limit) |
        (df[feature] > upper_limit)
    ]
    
    print("\nFeature:", feature)
    print("Number of outliers:", len(outliers))
```

```python
print("""
Inference:

1. The dataset contains four numerical features and one categorical feature.

2. Histograms show:
   - Petal_Length and Petal_Width have clear grouped distributions.
   - Sepal features are more normally distributed.

3. Boxplots indicate:
   - Sepal_Width contains some outliers.
   - Other features have fewer or no significant outliers.

4. Petal measurements provide better separation between flower species.
""")
```
