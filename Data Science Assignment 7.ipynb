```python
import nltk
import pandas as pd

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
```

```python
document = """
Text Analytics is the process of converting unstructured text data
into meaningful insights. Natural Language Processing helps computers
understand human language.
"""

print(document)
```

```python
# Tokenization

tokens = word_tokenize(document)

print("Tokens:\n")
print(tokens)
```

```python
import nltk

nltk.download('averaged_perceptron_tagger_eng')
```

```python
# POS Tagging

pos_tags = pos_tag(tokens)

print("POS Tags:\n")

for word, tag in pos_tags:
    print(word, ":", tag)
```

```python
# Remove stopwords

stop_words = set(stopwords.words('english'))

filtered_words = [
    word for word in tokens
    if word.lower() not in stop_words
]

print("After Stopword Removal:\n")
print(filtered_words)
```

```python
# Stemming

ps = PorterStemmer()

stemmed_words = [ps.stem(word) for word in filtered_words]

print("Stemmed Words:\n")
print(stemmed_words)
```

```python
# Lemmatization

lemmatizer = WordNetLemmatizer()

lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

print("Lemmatized Words:\n")
print(lemmatized_words)
```

```python
documents = [
    "Machine learning is fascinating",
    "Text analytics uses natural language processing",
    "Machine learning and NLP are important fields"
]

print(documents)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create CountVectorizer object

cv = CountVectorizer()

# Transform documents

tf_matrix = cv.fit_transform(documents)

# Convert to DataFrame

tf_df = pd.DataFrame(
    tf_matrix.toarray(),
    columns=cv.get_feature_names_out()
)

print("Term Frequency Matrix:\n")

print(tf_df)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF Vectorizer

tfidf = TfidfVectorizer()

# Transform documents

tfidf_matrix = tfidf.fit_transform(documents)

# Convert to DataFrame

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

print("TF-IDF Matrix:\n")

print(tfidf_df)
```
